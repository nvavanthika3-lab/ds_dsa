```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
```

```python
n1=Node("anandhu")
n2=Node("avanthika")
n3=Node("karthi")
```

```python
n1.next=n2
n1.prev=None
```

```python
n2.next=n3
n2.prev=n1
```

```python
n3.next=None
n3.prev=n2
```

```python
cur=n1
while cur:
    print(cur.data)
    cur=cur.next
```

```python
print(n1.next.next.data)
```

```python
print(n2.prev.next.data)
```
