```python
class Order:
    def __init__(self,id,date,price,qty):
        self.id=id
        self.date=date
        self.price=price
        self.qty=qty
    def display(self):
        print(f"Orderid:{self.id} OrderDate:{self.date} OrderPrice:{self.price} Orderqty:{self.qty}")
```

```python
o1=Order(100,"10-10-2026",2200,2)
o2=Order(101,"15-07-2026",25000,3)
o3=Order(300,"04-04-2026",10000,4)
o4=Order(400,"04-10-2025",5000,2)
o5=Order(102,"10-09-2026",2500,2)
```

```python
o1.display()
```

```python
o2.display()
```

```python
o3.display()
```

```python
o4.display()
```

```python
from collections import deque
```

```python
order_queue=deque()
```

```python
order_queue.append(o1)
order_queue.append(o2)
order_queue.append(o3)
order_queue.append(o4)
order_queue.append(o5)
```

```python
while order_queue:
    cur=order_queue.popleft()
    print("processing order")
    cur.display()
```

```python
class Patient:
    def __init__ (self,id,name,department,type):
        self.id=id
        self.name=name
        self.department=department
        self.type=type
    def patient_info(self):
                print(f"Patientid:{self.id} Patientname:{self.name} patientdepartment:{self.department} patienttype:{self.type}")
```

```python
p1=Patient(1,"kevin","xray","indor")
p2=Patient(2,"anandhu","ent","indor")
p3=Patient(3,"adhi","emergency","outdoor")
p4=Patient(4,"kuttu","casuality","outdoor")
p5=Patient(5,"avanthika","skin","indor")
```

```python
p1.patient_info()
```

```python
import heapq
```

```python
patients=[]
```

```python
heapq.heappush(patients,(5,p1))
heapq.heappush(patients,(2,p2))
heapq.heappush(patients,(1,p3))
heapq.heappush(patients,(3,p4))
heapq.heappush(patients,(4,p5))

```

```python
while patients:
    pr,p=heapq.heappop(patients)
    p.patient_info()
```
