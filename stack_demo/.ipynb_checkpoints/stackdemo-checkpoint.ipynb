```python
class Stack:
    def __init__(self):
        self.data=[]
    def push(self,item):
        self.data.append(item)
    def display(self):
        for i in range(-1,-len(self.data)+1,-1):
            print(self.data[i],sep='')
    def pop(self):
        if self.is_empty()!=True:
            return self.data.pop()
        else:
            print("Stack is empty")
    def is_empty(self):
        if len(self.data)==0:
            return True
```

```python
s=Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.push(50)
```

```python
print(s.data)
```

```python
s.display()
```
